```python
## TASK-10

import pandas as pd
import numpy as np

# -----------------------------------------------------------
# 1. Creating a MultiIndex Series in Different Ways
# -----------------------------------------------------------

arrays = [
    ["India", "India", "Australia", "Australia"],
    ["Virat", "Rohit", "Smith", "Warner"]
]
index = pd.MultiIndex.from_arrays(arrays, names=("Team", "Player"))
multi_s = pd.Series([82, 64, 95, 78], index=index)
print("MultiIndex Series from arrays:\n", multi_s, "\n")

tuples = [
    ("India", "Virat"), ("India", "Rohit"),
    ("Australia", "Smith"), ("Australia", "Warner")
]
index2 = pd.MultiIndex.from_tuples(tuples, names=("Team", "Player"))
multi_s2 = pd.Series([75, 68, 88, 90], index=index2)
print("MultiIndex Series from tuples:\n", multi_s2, "\n")

iterables = [["India", "Australia"], ["Virat", "Rohit"]]
index3 = pd.MultiIndex.from_product(iterables, names=("Team", "Player"))
multi_s3 = pd.Series(np.random.randint(40, 120, size=4), index=index3)
print("MultiIndex Series from product:\n", multi_s3, "\n")

# -----------------------------------------------------------
# 2. Accessing and Indexing
# -----------------------------------------------------------

print("Access all players from Team 'India':\n", multi_s.loc["India"], "\n")
print("Access specific element (Australia, Warner):\n", multi_s.loc[("Australia", "Warner")], "\n")

# -----------------------------------------------------------
# 3. Slicing in MultiIndex
# -----------------------------------------------------------

multi_s = multi_s.sort_index()  # prevent UnsortedIndexError
print("Slicing from India to Australia:\n", multi_s.loc["India":"Australia"], "\n")
print("Partial slice for all 'Virat':\n", multi_s.loc[:, "Virat"], "\n")

# -----------------------------------------------------------
# 4. Swapping and Reordering Levels
# -----------------------------------------------------------

print("Swapping levels:\n", multi_s.swaplevel(), "\n")
print("Reordering levels:\n", multi_s3.reorder_levels(["Player", "Team"]), "\n")

# -----------------------------------------------------------
# 5. Passing List of Arrays directly to Series / DataFrame
# -----------------------------------------------------------

multi_s_auto = pd.Series(
    np.random.randint(50, 120, size=4),
    index=pd.MultiIndex.from_arrays([["India", "India", "Australia", "Australia"],
                                     ["Virat", "Rohit", "Smith", "Warner"]])
)
print("MultiIndex Series constructed automatically:\n", multi_s_auto, "\n")

df_auto = pd.DataFrame(
    np.random.randint(40, 120, size=(4, 2)),
    index=pd.MultiIndex.from_arrays([["Match1", "Match1", "Match2", "Match2"],
                                     ["India", "Australia", "India", "Australia"]]),
    columns=["Runs", "StrikeRate"]
)
print("DataFrame with MultiIndex automatically:\n", df_auto, "\n")

# -----------------------------------------------------------
# 6. Data Alignment and Reindexing
# -----------------------------------------------------------

df = pd.DataFrame({
    "Virat": [85, 90, 95, 80],
    "Rohit": [82, 88, 92, 84]
}, index=pd.MultiIndex.from_arrays([["India", "India", "Australia", "Australia"],
                                    ["Innings1", "Innings2", "Innings1", "Innings2"]]))
print("Original DataFrame:\n", df, "\n")

mean_by_team = df.groupby(level=0).mean()
print("Mean by team:\n", mean_by_team, "\n")

aligned = mean_by_team.reindex(df.index, level=0)
print("Reindexed to align with original index:\n", aligned, "\n")

# -----------------------------------------------------------
# 7. Using xs() for Cross-Section
# -----------------------------------------------------------

print("Cross-section for 'Innings2':\n", df.xs("Innings2", level=1), "\n")

# -----------------------------------------------------------
# 8. Sorting and Removing Unused Levels
# -----------------------------------------------------------

unsorted = multi_s_auto.sample(frac=1)   # Shuffle randomly
print("Unsorted MultiIndex Series:\n", unsorted, "\n")
print("Sorted by index:\n", unsorted.sort_index(), "\n")

sub_df = df_auto[["Runs"]]   # Select one column
print("Unused levels before removing:\n", sub_df.columns, "\n")

if isinstance(sub_df.columns, pd.MultiIndex):
    sub_df.columns = sub_df.columns.remove_unused_levels()
    print("After remove_unused_levels:\n", sub_df.columns.levels, "\n")
else:
    print("Columns are not MultiIndex — no unused levels to remove.\n")
```

```python
## TASK-26

# Program 7: Sorting and Grouping in DataFrame 
import pandas as pd

# Cricket player stats
data = {
    "Player": ["Virat", "Rohit", "Jadeja", "Bumrah", "Dhawan"],  # Player names
    "Age": [34, 31, 30, 29, 33],                                 # Player ages
    "Runs": [75, 88, 60, 45, 90]                                  # Runs scored in last match
}
df = pd.DataFrame(data)
print("Original DataFrame (Player stats):\n", df, "\n")

# Sorting by Runs scored
print("Sorted by Runs:\n", df.sort_values(by="Runs"), "\n")

# Grouping by Age and computing average runs
grouped = df.groupby("Age")["Runs"].mean()
print("Average runs grouped by Age:\n", grouped, "\n")
```

```python
## TASK-27

# Grouping Example: Series 
import pandas as pd

# Runs scored by players
runs = pd.Series([75, 88, 60, 45],
                 index=["Virat", "Rohit", "Jadeja", "Bumrah"])

# Player roles (Batsman/Bowler)
role = pd.Series(["Batsman", "Batsman", "Allrounder", "Bowler"],
                 index=["Virat", "Rohit", "Jadeja", "Bumrah"])

# Grouping by role (no aggregation yet)
grouped_series = runs.groupby(role)
print("Grouped Series by Player Role (just groups, no aggregation):\n", grouped_series, "\n")

# Accessing a specific group: Batsmen
print("Batsmen group:\n", grouped_series.get_group("Batsman"), "\n")
```

```python
## TASK-28

# Aggregation Example: Series 
# Aggregation on grouped Series to compute stats per player role

# Compute mean runs per role
aggregated_series = grouped_series.mean()
print("Aggregated Series (mean runs per player role):\n", aggregated_series, "\n")

# Other aggregations
print("Total runs per role:\n", grouped_series.sum(), "\n")
print("Maximum runs scored by a player in each role:\n", grouped_series.max(), "\n")
```

```python
## TASK-29

# Example: Grouping and Aggregating for Series / MultiLevel Series 
import pandas as pd

# Single Series: Runs scored by players
runs = pd.Series([75, 88, 60, 45],
                 index=["Virat", "Rohit", "Jadeja", "Bumrah"])
role = pd.Series(["Batsman", "Batsman", "Allrounder", "Bowler"],
                 index=["Virat", "Rohit", "Jadeja", "Bumrah"])

# Grouping runs by player role and calculating mean
grouped_series = runs.groupby(role).mean()
print("Mean runs by player role (Series):\n", grouped_series, "\n")

# MultiLevel Series: Team and Player
arrays = [
    ["TeamA", "TeamA", "TeamB", "TeamB"],
    ["Virat", "Rohit", "Jadeja", "Bumrah"]
]
index = pd.MultiIndex.from_arrays(arrays, names=("Team", "Player"))
multi_s = pd.Series([75, 88, 60, 45], index=index)

# Group by first level (Team) and compute mean runs
grouped_multi = multi_s.groupby(level=0).mean()
print("Mean runs by team (MultiLevel Series):\n", grouped_multi, "\n")
```

```python
## TASK-31

# Aggregating DataFrame Example 
# Aggregation on grouped DataFrame to compute stats per player role
import pandas as pd

# DataFrame: Player stats
df = pd.DataFrame({
    "Role": ["Batsman", "Batsman", "Allrounder", "Bowler"],  # Player roles
    "Player": ["Virat", "Rohit", "Jadeja", "Bumrah"],        # Player names
    "Runs": [75, 88, 60, 45]                                  # Runs scored
})

# Grouping by Role
grouped_df = df.groupby("Role")

# Compute mean runs per role
aggregated_df = grouped_df["Runs"].mean()
print("Aggregated DataFrame (mean runs per role):\n", aggregated_df, "\n")

# Multiple aggregations: mean, sum, max runs per role
multi_agg_df = grouped_df.agg({
    "Runs": ["mean", "sum", "max"]
})
print("Aggregated DataFrame (multiple stats per role):\n", multi_agg_df, "\n")
```

```python
## TASK-32

import pandas as pd

# ------------------------------------------
# Sample DataFrame: Cricket player stats
# ------------------------------------------
df = pd.DataFrame({
    "Team": ["TeamA", "TeamA", "TeamB", "TeamB", "TeamA", "TeamB"],  # Teams
    "Player": ["Virat", "Rohit", "Jadeja", "Bumrah", "Dhawan", "Hardik"],  # Players
    "Runs": [75, 88, 60, 45, 90, 70],        # Runs scored in recent matches
    "Balls": [50, 45, 40, 30, 55, 50]        # Balls faced
})

print("Original DataFrame (Player stats):\n", df, "\n")

# ------------------------------------------
# Grouping by Team (no aggregation yet)
# ------------------------------------------
grouped = df.groupby("Team")
print("Groups formed by Team:\n", grouped.groups, "\n")

# Access a specific group: TeamA
print("TeamA group:\n", grouped.get_group("TeamA"), "\n")

# ------------------------------------------
# Aggregating after grouping
# ------------------------------------------
# 1. Single aggregation function: mean runs per team
mean_runs = grouped["Runs"].mean()
print("Mean Runs per Team:\n", mean_runs, "\n")

# 2. Multiple aggregation functions: runs and balls stats
agg_stats = grouped.agg({
    "Runs": ["mean", "max", "min"],
    "Balls": ["sum", "mean"]
})
print("Aggregated stats per Team:\n", agg_stats, "\n")

# ------------------------------------------
# Optional: filtering groups (e.g., teams with mean runs > 70)
# ------------------------------------------
high_scoring_team = grouped.filter(lambda x: x["Runs"].mean() > 70)
print("Teams with mean runs > 70:\n", high_scoring_team, "\n")
```

```python
## TASK-35

import pandas as pd

# ------------------------------------------
# 1. Series Example: Runs scored by players
# ------------------------------------------
runs_series = pd.Series([75, 88, 60, 45],
                        index=["Virat", "Rohit", "Jadeja", "Bumrah"])
role_series = pd.Series(["Batsman", "Batsman", "Allrounder", "Bowler"],
                        index=["Virat", "Rohit", "Jadeja", "Bumrah"])

# Grouping (creates groups by role, no aggregation yet)
grouped_series = runs_series.groupby(role_series)
print("Grouped Series by Player Role (no aggregation):")
for role, group in grouped_series:
    print(f"{role}: {group.values}")
print()

# Aggregating (mean runs per role)
aggregated_series = grouped_series.mean()
print("Aggregated Series (mean runs per role):")
print(aggregated_series)
print("\n" + "="*50 + "\n")

# ------------------------------------------
# 2. DataFrame Example: Player stats
# ------------------------------------------
df = pd.DataFrame({
    "Role": ["Batsman", "Batsman", "Allrounder", "Bowler"],
    "Player": ["Virat", "Rohit", "Jadeja", "Bumrah"],
    "Runs": [75, 88, 60, 45],
    "Balls": [50, 45, 40, 30]  # Balls faced
})

# Grouping by Role (creates groups)
grouped_df = df.groupby("Role")
print("Grouped DataFrame by Player Role (no aggregation):")
for role, group in grouped_df:
    print(f"{role} group:\n{group}\n")

# Aggregating (mean and sum for numeric columns)
aggregated_df = grouped_df.agg({
    "Runs": ["mean", "sum"],
    "Balls": ["mean", "sum"]
})
print("Aggregated DataFrame (Runs and Balls stats per Role):")
print(aggregated_df)
```

```python
## TASK-36

# Merging DataFrames Example 
import pandas as pd

# DataFrame 1: Player teams
df1 = pd.DataFrame({
    "Player": ["Virat", "Rohit", "Jadeja", "Bumrah"],
    "Team": ["TeamA", "TeamB", "TeamA", "TeamB"]
})

# DataFrame 2: Player runs in recent match
df2 = pd.DataFrame({
    "Player": ["Virat", "Rohit", "Jadeja", "Dhawan"],
    "Runs": [75, 88, 60, 90]
})

# Inner merge: only players present in both DataFrames
inner_merge = pd.merge(df1, df2, on="Player", how="inner")
print("Inner Merge (Players in both Team & Runs DataFrames):\n", inner_merge, "\n")

# Outer merge: all players from both DataFrames
outer_merge = pd.merge(df1, df2, on="Player", how="outer")
print("Outer Merge (All players from both DataFrames):\n", outer_merge, "\n")
```

```python
## TASK-37

# Merging DataFrames on different keys 
import pandas as pd

# DataFrame 1: Player IDs and Names
df1 = pd.DataFrame({
    "PlayerID": [1, 2, 3],
    "Player": ["Virat", "Rohit", "Jadeja"]
})

# DataFrame 2: Player scores with a different key column name
df2 = pd.DataFrame({
    "PID": [2, 3, 4],
    "Runs": [88, 60, 90]
})

# Merge using different column names
merged_df = pd.merge(df1, df2, left_on="PlayerID", right_on="PID", how="inner")
print("Merge on different keys (Players present in both DataFrames):\n", merged_df)
```

```python
## TASK-38

# Handling overlapping column names 
import pandas as pd

# DataFrame 1: Player runs in previous match
df1 = pd.DataFrame({
    "Player": ["Virat", "Rohit"],
    "Runs": [75, 88]
})

# DataFrame 2: Player runs in current match
df2 = pd.DataFrame({
    "Player": ["Rohit", "Jadeja"],
    "Runs": [90, 60]
})

# Merge with custom suffixes for overlapping column names
merged_df = pd.merge(df1, df2, on="Player", how="outer", suffixes=("_Prev", "_Curr"))
print("Merge with custom suffixes (Previous and Current Runs):\n", merged_df)
```

```python
## TASK-40

# Summary Tables 
import pandas as pd

# DataFrame: Player runs by team and role
df = pd.DataFrame({
    "Role": ["Batsman", "Batsman", "Bowler", "Bowler"],  # Player roles
    "Team": ["TeamA", "TeamB", "TeamA", "TeamB"],        # Teams
    "Runs": [75, 88, 60, 45]                              # Runs scored
})

# Create pivot table: average runs by Role and Team
summary = df.pivot_table(values="Runs", index="Role", columns="Team", aggfunc="mean")

print("Pivot Table (Average Runs by Role and Team):\n", summary)
```

```python
## TASK-41

# Summary Table / Pivot Table 
import pandas as pd

# DataFrame: Player runs by Role and Team
df = pd.DataFrame({
    "Role": ["Batsman", "Batsman", "Bowler", "Bowler", "Batsman", "Bowler"],  # Player roles
    "Team": ["TeamA", "TeamB", "TeamA", "TeamB", "TeamA", "TeamB"],            # Teams
    "Runs": [75, 88, 60, 45, 90, 70]                                           # Runs scored
})

# Pivot table: average runs by Role and Team
summary_table = df.pivot_table(values="Runs", index="Role", columns="Team", aggfunc="mean")

print("Summary Table (Average Runs by Role and Team):\n", summary_table)
```

```python
## TASK-42

# Enhanced Pivot Table with Multiple Aggregations 
import pandas as pd

# DataFrame: Player runs by Role and Team
df = pd.DataFrame({
    "Role": ["Batsman", "Batsman", "Bowler", "Bowler", "Batsman", "Bowler"],
    "Team": ["TeamA", "TeamB", "TeamA", "TeamB", "TeamA", "TeamB"],
    "Runs": [75, 88, 60, 45, 90, 70]
})

# Pivot table: calculate mean, sum, and max runs by Role and Team
summary_table = df.pivot_table(
    values="Runs",
    index="Role",
    columns="Team",
    aggfunc=["mean", "sum", "max"]
)

print("Enhanced Pivot Table (Runs stats by Role and Team):\n", summary_table)
```

```python
## TASK-43

# Handling Missing Data in Pivot Tables 
import pandas as pd

# DataFrame: Player runs by Role and Team (some combinations missing)
df = pd.DataFrame({
    "Role": ["Batsman", "Batsman", "Bowler", "Bowler", "Batsman"],
    "Team": ["TeamA", "TeamB", "TeamA", "TeamB", "TeamA"],
    "Runs": [75, 88, 60, 45, 90]
})

# Pivot table: average runs by Role and Team, fill missing with 0
summary_table_fill = df.pivot_table(
    values="Runs",
    index="Role",
    columns="Team",
    aggfunc="mean",
    fill_value=0
)

print("Pivot Table with Missing Values Filled (Runs):\n", summary_table_fill)
```

```python
## TASK-44

# Grouping vs Pivot Tables 
import pandas as pd

# DataFrame: Player runs by Role and Team
df = pd.DataFrame({
    "Role": ["Batsman", "Batsman", "Bowler", "Bowler", "Batsman", "Bowler"],
    "Team": ["TeamA", "TeamB", "TeamA", "TeamB", "TeamA", "TeamB"],
    "Runs": [75, 88, 60, 45, 90, 70]
})

# Equivalent of pivot table using groupby + unstack
grouped_equiv = df.groupby(["Role", "Team"])["Runs"].mean().unstack()

print("Equivalent Pivot Table using groupby + unstack:\n", grouped_equiv)
```
